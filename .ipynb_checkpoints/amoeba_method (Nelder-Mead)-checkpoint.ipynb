```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
def f(x):
    y = (np.cos(30*x) + np.sin(15*x)/3 - np.cos(5*x)) 
    return y

f_v = np.vectorize(f)
```

```python
X = np.linspace(-1, 1.1, num = 1000)
plt.plot(X, f_v(X))
```

```python
from scipy import optimize
```

```python
x0 = [0.0]
minimizer_kwargs = { "method": "Nelder-Mead" }
x1 = optimize.basinhopping(f, x0, minimizer_kwargs=minimizer_kwargs, niter = 200)

print (x1.fun)
```
