```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from math import *
```

```python
def f(x):
    y = sin(x/5.0) * exp(x/10.0) + 5 * exp(-x/2.0)
    return y
```

```python
O_x = np.arange(1, 30, 0.1)
O_y = [f(x) for x in O_x]

plt.plot(O_x, O_y)
```

## 1. Minimization of a smooth function

```python
# Reaching the first minimum

x0 = [2.0]
minimizer_kwargs = { "method": "BFGS" } #using Broyden–Fletcher–Goldfarb–Shanno algorithm
x1 = optimize.basinhopping(f, x0, minimizer_kwargs=minimizer_kwargs, niter = 200)

print (round(x1.fun, 2))
```

```python
# Reaching the second minimum

x0 = [30.0]
minimizer_kwargs = { "method": "BFGS" }
x2 = optimize.basinhopping(f, x0, minimizer_kwargs=minimizer_kwargs, niter = 200)

print(round(x2.fun, 2))
```

## 2. Global optimization 
###    Differential Evolution

```python
bounds = [(1, 30)]
result = optimize.differential_evolution(f, bounds)

print("x_min = {}, y_min = {}".format(result.x[0], round(result.fun, 2)))
```

```python
plt.plot(O_x, O_y)
plt.scatter(result.x[0], result.fun, marker = 'x')
```

## 3. Minimization of a non-smooth function

```python
def f(x):
    y = sin(x/5) * exp(x/10) + 5 * exp(-x/2)
    return y


def h(x):
    return int(f(x))

h_v = np.vectorize(h)
```

```python
Ox = np.linspace(1, 30, num = 100)
plt.plot(Ox, h_v(Ox))
```

```python
x0 = [30]
minimizer_kwargs = { "method": "BFGS" }
x1 = optimize.basinhopping(h, x0, minimizer_kwargs=minimizer_kwargs, niter = 1)

print (x1.fun)
```

```python
bounds = [(1, 30)]
result = optimize.differential_evolution(h, bounds)

print("x_min = {}, y_min = {}".format(result.x[0], result.fun))
```
